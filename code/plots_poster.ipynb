```python
from sklearn.preprocessing import StandardScaler
import seaborn as sns 
```

```python
sns.set_style("dark")
tips = sns.load_dataset("tips")[["total_bill", "tip"]]

scaler = StandardScaler()
data = scaler.fit_transform(tips)

fig, ax = plt.subplots()
sns.scatterplot(x=data[0:10, 0], y=data[0:10, 1])
fig.savefig("plots/example_embedding1.pdf")

fig, ax = plt.subplots()
sns.scatterplot(x=data[10:20, 0], y=data[10:20, 1])
fig.savefig("plots/example_embedding2.pdf")

fig, ax = plt.subplots()
sns.scatterplot(x=data[20:30, 0], y=data[20:30, 1])
fig.savefig("plots/example_embedding3.pdf")
```
